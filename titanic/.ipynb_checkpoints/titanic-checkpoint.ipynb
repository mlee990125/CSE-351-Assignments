```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import random
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('titanic/train.csv')
df2 = pd.read_csv('titanic/test.csv')
print(df.isnull().sum())
print(df2.isnull().sum())
# First of all, we just dropped the column with the passenger ID since it can be replaced by the index number
# since cabin has too many missing values and is not able to predict, we have decided to delete the coulumn.
# Since name has nothing to deal with the survival, we dropped it from the dataframe.
# Ticket number was differenct from one another except the passengers are family. It is almost the same as 
# passenger ID since it is unique for every people unless they are not family. So we dropped it.

df = df.drop(["PassengerId","Cabin","Name","Ticket"],axis=1)
df2 = df2.drop(["PassengerId","Cabin","Name","Ticket"],axis=1)

### Fill in the missing values.
# For the age, we think that there should be a correlation between the social status and the age.
# So we tried to interpolate the value of age according to the correlated attributes with the interpolation.
# We use forward selection method to make sure that the ages are filled correctly with the related attributes.

df["Age"].interpolate(method ='linear', limit_direction ='forward', inplace=True)
df2["Age"].interpolate(method ='linear', limit_direction ='forward', inplace=True)

# For Embarked we will just replace the value randomly. Since there are only two missing values.
# For Fare in test data, we will just replcae the value with the mean of the fares.

df["Embarked"].fillna(random.choice(['S','Q','C']),inplace=True)
df2["Fare"].fillna(df2["Fare"].mean(),inplace=True)
print(df.isnull().sum())
print(df2.isnull().sum())

# Now all the missing values are filled.
### We think removing outliers for this data is not a good way to preprocess the data since most attributes are
### qualitative values. For quantitative attribtues, they are all discrete values without a wide range. So the only
### field that may contain outliers is age. For age, it is not appropriate just deleting the record with high age
### since it is not an error.

# To make the analysis easier, we divided the dataframe of the train dataset into quantitative values and qualitative values

# First of all groupby some categorical values and count the survival rate of the people according to the value of
# the categorical value.
# The result shows that the Sex has the biggest difference in survival rate with the biggest difference.
# The attribute Emabarked shows that the value with C has majority of people survived.
# The attribute Pclass shows that most people from the class 1 have survived.

print(df["Survived"].mean())
print(df.groupby("Sex")["Survived"].mean())
print(df.groupby("Embarked")["Survived"].mean())
print(df.groupby("Pclass")["Survived"].mean())

df[['Survived','Pclass','Sex','Embarked']] = df[['Survived','Pclass','Sex','Embarked']].astype("category")
# male=0 female=1
df['Sex'] = pd.factorize(df['Sex'])[0]
# S=0 C=1 Q=2
df['Embarked'] = pd.factorize(df['Embarked'])[0]

df2['Sex'] = df2['Sex'].map({
    'male': 0, 
    'female': 1
})

df2['Embarked'] = df2['Embarked'].map({
    'S': 0, 
    'C': 1,
    'Q': 2
})

```

```python
### TO FIND THE RELATIONSHIP BETWEEN SOCIO-ECONOMIC STATUS OF A PERSON, THERE IS ONLY ONE FEILD THAT INDICATES THE SOCIAL-ECONOMIC STATUS OF AN INDIVIDUAL.
### So we plotted each field according to the Pclass value.


plt.figure(figsize=(24,8))
plt.subplot(2,4,1)
sns.histplot(data=df, x="Age", hue="Pclass")

plt.subplot(2,4,2)
sns.histplot(data=df, x="SibSp", hue="Pclass")

plt.subplot(2,4,3)
sns.histplot(data= df, x="Parch", hue="Pclass")

plt.subplot(2,4,4)
sns.histplot(data=df, x="Fare", hue="Pclass")

plt.subplot(2,4,5)
sns.histplot(data=df, x="Embarked", hue="Pclass")

plt.subplot(2,4,6)
sns.histplot(data=df, x="Sex", hue="Pclass")

df['Survived'] = df['Survived'].cat.as_ordered()

plt.subplot(2,4,7)
sns.histplot(data=df, x="Survived", hue="Pclass")

# From the grid and the histogram given from the data, most of data are skewed to the right. Therefore, it is difficult
# to tell there is a certain relationship between pair of variables. However, the attribute fare clearly separates
# Pclass 1 people from the other two classes accroding to the plot given. Embarked field also shows that there are many
# Pclass 1 people in Cherbourg while Queenstown has dominant Pclass 3 people. Southampton is the town with most people
# Survived field also has some abnormal feature. Since Pclass 3 is the dominant number of people, those who survived
# is supposed to be also donimated by the class 3 people. However, the class that most people survived was Pclass1.

#sns.pairplot(df,hue = "Survived")

```

```python
### Distribution of the survival victims in relation to age, gender, Sibsp, Parch, Fare, Embarkded, and Pclass.

plt.figure(figsize=(24,8))

plt.subplot(2,4,1)
sns.histplot(data=df, x="Age", hue="Survived")

plt.subplot(2,4,2)
sns.histplot(data=df, x="SibSp", hue="Survived")

plt.subplot(2,4,3)
sns.histplot(data= df, x="Parch", hue="Survived")

plt.subplot(2,4,4)
sns.histplot(data=df, x="Fare", hue="Survived")

plt.subplot(2,4,5)
sns.histplot(data=df, x="Embarked", hue="Survived")

plt.subplot(2,4,6)
sns.histplot(data=df, x="Sex", hue="Survived")

df['Pclass'] = df['Pclass'].cat.as_ordered()

plt.subplot(2,4,7)
sns.histplot(data=df, x="Pclass", hue="Survived")

# From the histogram below, it is shown that people from Emarked value C, Sex value female, age value less than 8,
# Fare with higher value, or Pclass 1 have majority of people survived. Considering that majority of people were 
# dead, more than half is an abnormally high probability to survive.
```

```python
# Calculate the correlation matrix first to check if there is highly correlated fields.

plt.subplots(figsize=(10,10))
sns.heatmap(df[['Age','SibSp','Parch','Fare','Survived']] .astype('float').corr(),annot=True)

# The correlation matrix below does not show any pair of attributes with high correlation. Sibsp and Parch is the
# the only pair with some linear correlation. Sibsp has relatively high negative correlation with Age. However, it
# is still not enough to say that there is a strong relationship between those two pairs with 0.21 ablsolute value.

### From the information gathered from the above analysis, the most important feature that is directly related to the
### survival rate of the people is sex since it has the most obvious and clear result with the highest survival rate.
### It has also the biggest difference between survival rate with its counter value.
### Through pairing, I was able to find there are some categorical values that is related to the survival rate.
```

```python
### MODEL 1 : Logistic Regression
# The first model that we are going to build our classifier is the logistic regression model. The reason we chose
# this model is that we have learned this model during our class. It has a strength with handling both qualitative
# and quantitative variables as independent variables (or x values). It is also good at handling dependent variable
# with binary class. That means true or false for survival. Since we consider Survived as our class attribute.

import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import cross_validate


test_y = pd.read_csv('titanic/test_Y.csv')
test_y = test_y["Survived"]
log = LogisticRegression(max_iter=1000)

train_x = df.drop(columns=["Survived"])
train_y = df["Survived"]
log.fit(train_x,train_y)

predY1 = log.predict(df2)

print("Logistic Regression Result")

# print the confusion matrix
print(confusion_matrix(test_y, predY1))

# print Precision, Recall, F1 score of the model
print(classification_report(test_y,predY1))

from sklearn.model_selection import cross_val_score,KFold
# For our cross-validation, we chose 10-fold cross validation. Leave one and out is too expensive to calculate.
# We also do not want any intersection between folds to make sure the model maximize the given information correctly.
# We found the best cross-validation method that meets these condition is k-fold cross validation. We improvised the
# number k to 10 since most of the expert set is one of the most common value of k in k-fold cross validation.
# Merge the test and train data set to do cross validation since it automatically divides into two classes.

temp = pd.concat([test_y,df2],axis=1)
temp = pd.concat([df,temp],axis=0,ignore_index=True)

temp_x = temp.drop(columns=["Survived"])
temp_y = temp["Survived"]

# 10-fold validation result

kf=KFold(n_splits=5)

print("BELOW IS THE AVERAGE SCORE FOR THE 10-FOLD CROSS VALIDATION OF LOGISITC REGRESSION")

score = cross_val_score(log,temp_x,temp_y,cv=kf, scoring='accuracy')
print("Cross Validation Accuracy Scores are {}".format(score))
print("Average Cross Validation Accuracy score :{}".format(score.mean()))

score = cross_val_score(log,temp_x,temp_y,cv=kf, scoring='precision')
print("Cross Validation Precision Scores are {}".format(score))
print("Average Cross Validation Precision score :{}".format(score.mean()))

score = cross_val_score(log,temp_x,temp_y,cv=kf, scoring='recall')
print("Cross Validation Recall Scores are {}".format(score))
print("Average Cross Validation Recall score :{}".format(score.mean()))

score = cross_val_score(log,temp_x,temp_y,cv=kf, scoring='f1')
print("Cross Validation F1 Scores are {}".format(score))
print("Average Cross Validation F1 score :{}".format(score.mean()))

### From the original accuracy score of the logisitc regression, all the values of metrics have decreased
### Therefore, 10-fold cross validation did not improve the performance of the model.


```

```python
### MODEL 2 : Decision Tree
# The second model that we chose is a decision tree. It is an algorithm that uses categorizing the records according to
# the class attribute. It is a unsupervised descripitive classification method. It build a tree using the traing data.
# When all the records in a given subtree have the same classification label, then the subtree becomes leaf. 
# If there is no record availble to proceed further, or all the class attributes are used to build a path from the
# root, then majority voting occurs. If there is two or more class value in a given tree with terminating condition,
# it stops iterating count the majority of the class labels. The class label for that leaf becomes the majority class.

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()

dtc.fit(train_x,train_y)

print("Decision Tree Classifier Result")

predY3 = dtc.predict(df2)

# print the confusion matrix
print(confusion_matrix(test_y, predY3))

# print Precision, Recall, F1 score of the model
print(classification_report(test_y,predY3))

# 10-fold validation result

print("BELOW IS THE AVERAGE SCORE FOR THE 10-FOLD CROSS VALIDATION OF DECISION TREE CLASSIFIER")

score = cross_val_score(dtc,temp_x,temp_y,cv=kf, scoring='accuracy')
print("Cross Validation Accuracy Scores are {}".format(score))
print("Average Cross Validation Accuracy score :{}".format(score.mean()))

score = cross_val_score(dtc,temp_x,temp_y,cv=kf, scoring='precision')
print("Cross Validation Precision Scores are {}".format(score))
print("Average Cross Validation Precision score :{}".format(score.mean()))

score = cross_val_score(dtc,temp_x,temp_y,cv=kf, scoring='recall')
print("Cross Validation Recall Scores are {}".format(score))
print("Average Cross Validation Recall score :{}".format(score.mean()))

score = cross_val_score(dtc,temp_x,temp_y,cv=kf, scoring='f1')
print("Cross Validation F1 Scores are {}".format(score))
print("Average Cross Validation F1 score :{}".format(score.mean()))

### From the original accuracy score of the decision tree classifier, all the values of metrics have decreased
### or maintained. Therefore, 10-fold cross validation did not improve the performance of the model.

```

```python
### MODEL 3 : Random Forest Classifier
# Random Forest is the machine learning method for classification, regression, and other tasks.
# It operates by constructing a multitude of decision trees at training time.
# For regression tasks, the mean or average prediction of the individual trees is returned.
# Random forests are a way of averaging multiple deep decision trees, trained on 
# different parts of the same training set, with the goal of reducing the variance.
# Random forests generate reasonable predictions across a wide range of data while requiring little configuration.

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc.fit(train_x,train_y)

print("Random Forest Classifier Result")

predY2 = rfc.predict(df2)

# print the confusion matrix
print(confusion_matrix(test_y, predY2))

# print Precision, Recall, F1 score of the model
print(classification_report(test_y,predY2))

# 10-fold validation result

print("BELOW IS THE AVERAGE SCORE FOR THE 10-FOLD CROSS VALIDATION OF RANDOM FOREST CLASSIFIER")

score = cross_val_score(rfc,temp_x,temp_y,cv=kf, scoring='accuracy')
print("Cross Validation Accuracy Scores are {}".format(score))
print("Average Cross Validation Accuracy score :{}".format(score.mean()))

score = cross_val_score(rfc,temp_x,temp_y,cv=kf, scoring='precision')
print("Cross Validation Precision Scores are {}".format(score))
print("Average Cross Validation Precision score :{}".format(score.mean()))

score = cross_val_score(rfc,temp_x,temp_y,cv=kf, scoring='recall')
print("Cross Validation Recall Scores are {}".format(score))
print("Average Cross Validation Recall score :{}".format(score.mean()))

score = cross_val_score(rfc,temp_x,temp_y,cv=kf, scoring='f1')
print("Cross Validation F1 Scores are {}".format(score))
print("Average Cross Validation F1 score :{}".format(score.mean()))

### From the original accuracy score of the decision tree classifier, most of the values of metrics have decreased.
### The only metric that had been increase from the previous model is the precision. However, it is almost the same
### as the previous precision while other metrics are decreased more severely. Therefore, 10-fold cross validation 
### did not improve the performance of the model.
```
